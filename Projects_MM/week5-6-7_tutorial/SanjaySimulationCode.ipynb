```python
import seaborn as sn
n=10000
mealCost=11
fixedCost=3995
mealsSold=np.random.normal(loc=3000,scale=1000,size=n)
revenuePerMeal=np.random.choice([20,18.5,16.5,15],size=n,p=[0.25,0.35,0.3,0.1])
laborCost=5040+(6850-5040)*np.random.random(size=n)
profit=mealsSold*(revenuePerMeal-mealCost)-laborCost-fixedCost
ax=sn.histplot(profit,kde=True,bins=50,stat="probability")
ax.set(xlabel='Monthly Earnings', ylabel='Probability')
```

```python
import seaborn as sn
n=30000
mealCost=11
fixedCost=3995
mealsSold=np.random.normal(loc=36000,scale=3464,size=n)
revenuePerMeal=np.random.choice([20,18.5,16.5,15],size=n,p=[0.25,0.35,0.3,0.1])
laborCost=5040+(6850-5040)*np.random.random(size=n)
profit=mealsSold*(revenuePerMeal-mealCost)-12*laborCost-12*fixedCost
salary1=profit
ax=sn.histplot(profit,kde=True,bins=50,stat="probability")
ax.set(xlabel='Yearly Earnings', ylabel='Probability')
```

```python
n=30000
mealCost=11
fixedCost=3995
mealsSold=np.random.normal(loc=3000,scale=1000,size=n*12).reshape(12,n)
revenuePerMeal=np.random.choice([20,18.5,16.5,15],size=n,p=[0.25,0.35,0.3,0.1])
laborCost=5040+(6850-5040)*np.random.random(size=n)
profit=mealsSold*(revenuePerMeal-mealCost)-laborCost-fixedCost
salary=np.where(profit<3000,3000,profit)
salary=np.where(salary<9000,salary,9000+0.1*(salary-9000))
salary=salary.sum(axis=0)
ax=sn.histplot([salary1,salary],color=['r','b'], alpha=0.8,kde=True,bins=50,stat="probability",legend=False)
ax.set(xlabel='Yearly Earnings', ylabel='Probability')
```
