```python
%pylab inline
from gurobipy import *
import random
```

```python
#Data
northTeams=["ANTF","CBLOA","CBSAL","LRNA","CQMB"]
centerTeams=["EVRT","UE","WDRS","PLTN","UCH","STGM","COLO","AUDAX","CATO","OHG"]
southTeams=["RNGS","UDC","CONCE","HCH","PMNTT"]
popularTeams=["COLO","CATO","UCH"]
strongTeams=popularTeams+["CBLOA"]
santiagoTeams=["CATO","AUDAX","PLTN","UCH","COLO","UE","STGM"]

teams=northTeams+centerTeams+southTeams
random.seed(1)         #fixing seed to make it reconstructible 
random.shuffle(teams) # randomizing order to assign to groups
groupA=teams[:5]
groupB=teams[5:10]
groupC=teams[10:15]
groupD=teams[15:]
groups={"A":groupA,"B":groupB,"C":groupC,"D":groupD}

gameDays=np.arange(1,20)
gameDays18=np.arange(1,19)
print ("Group A     ",groupA)
print ("Group B     ",groupB)
print ("Group C     ",groupC)
print ("Group D     ",groupD)
print ("gamedays             ",gameDays)
print ("gamedays until 18    ",gameDays18)
print("all teams",teams)
```

![title](vars.png)


```python
model=Model("League")
x=model.addVars(teams,teams,gameDays,vtype=GRB.BINARY,name="x")  
model.addConstrs(  ( x[i,i,k]==0  for i in teams for k in gameDays), name='dummy')  # remember to set x_{iik}=0
```

![title](obj.png)

```python
obj=0
for g in groups.keys():
    obj+=quicksum(k*x[i,j,k] for i in groups[g] for j in groups[g] for k in range(10,19))
    obj+=quicksum(15*x[i,j,19] for i in groups[g] for j in groups[g] )
model.setObjective(obj,sense=GRB.MAXIMIZE)
```

![title](CON1-3.png)

```python
#con1
model.addConstrs( 
    quicksum(x[i,j,k] + x[j,i,k] for k in gameDays)==1 
    for i in teams for j in teams if i!=j   ) 
#con2
model.addConstrs( quicksum(x[i,j,k]+x[j,i,k] for j in teams)==1 
                 for i in teams for k in gameDays)
#con3
model.addConstrs( quicksum(x[i,j,k] for j in teams for k in gameDays)>=9 for i in teams )
model.addConstrs( quicksum(x[i,j,k] for j in teams for k in gameDays)<=10 for i in teams )
```


![title](CON7.png)


```python

#7
for g in groups.keys():
    model.addConstrs(quicksum(x[i,j,k] for k in gameDays for j in groups[g])==2 for i in groups[g])

```

![title](Con9-10.png)

```python
#9
model.addConstrs(quicksum(x[j,i,k]+x[j,i,k+1] for j in teams)-
                 quicksum(x[j,i,k]+x[j,i,k+1] for j in northTeams)<=1 
                 for k in gameDays18 for i in northTeams )


model.addConstrs(quicksum(x[j,i,k]+x[j,i,k+1] for j in teams)-
                 quicksum(x[j,i,k]+x[j,i,k+1] for j in southTeams)<=1 for k in gameDays18 for i in southTeams )
#10
model.addConstrs(quicksum(x[j,i,k]+x[j,i,k+1] for j in teams)-
                 quicksum(x[j,i,k]+x[j,i,k+1] for j in centerTeams)<=1 for k in gameDays18 for i in centerTeams )
```

```python

#11
model.addConstrs(
    quicksum(x[h,"COLO",k]+x[h,"UCH",k] for k in gameDays)==1 for h in [l for l in teams if l not in ["COLO","UCH"] ] )
#12
model.addConstr(quicksum(x[i,j,k] for i in popularTeams for j in popularTeams for k in range(1,10))==0)
model.addConstr(quicksum(x[i,j,k] for i in popularTeams for j in popularTeams for k in range(17,20))==0)
#13
model.addConstr(quicksum(x["CATO","COLO",k]+x["UCH","COLO",k]-x["CATO","UCH",k]-x["COLO","UCH",k] for k in gameDays)==0)
#14
model.addConstrs(quicksum(x[i,j,k] for i in northTeams)+quicksum(x[i,h,k] for i in southTeams)<=1
                for k in gameDays for j in popularTeams for h in popularTeams if j!=h)
#16
model.addConstrs(quicksum( x[i,j,k] for j in strongTeams )<=1 for k in gameDays18 for i in teams)
#17
model.addConstr(quicksum(x["CBLOA",j,k]+x[j,"CBLOA",k] for j in popularTeams for k in range(1,6))==0)
model.addConstr(quicksum(x["CBLOA",j,19]+x[j,"CBLOA",19] for j in popularTeams )==0)
#20
model.addConstrs(quicksum(x[i,j,k] for i in santiagoTeams for j in teams)<=4 for k in gameDays)
model.addConstrs(quicksum(x[i,j,k] for i in santiagoTeams for j in teams)>=2 for k in gameDays)
#24
expr=0
for g in groups.keys():
    expr+=quicksum(x[i,j,19] for i in groups[g] for j in groups[g])
model.addConstr(expr<=3)
```
